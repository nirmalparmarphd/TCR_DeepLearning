```python
# droping Nan from dataset
clean_data = raw_data.dropna(axis=0)
clean_data.info()
```

```python
# data mapping
clean_data['Pair'].unique()
```

```python
# 'Al-Br-Cu' = 1
# 'Al-Air-Al' = 2
# 'Al-Air-Cu' = 3

clean_data.iloc[:,0] = clean_data.iloc[:,0].map({'Al-Br-Cu':1, 'Al-Air-Al':2, 'Al-Air-Cu':3})
clean_data.iloc[:,1] = clean_data.iloc[:,1].map({'Ambient':0, 'Pressure':1})
```

## Data Checkpoint

```python
clm = ['pairs', 'pressure', 'T1', 'T1i', 'T2i', 'T2', 'dT', 'dTi']
clean_data.columns = clm 
```

```python
processed_data = clean_data.copy()

## NOTE
# wop --> with out pressure
# wp --> with pressure

# data measured without pressure
processed_data_wop = processed_data[processed_data['pressure']==0]
# data measured with pressure
processed_data_wp = processed_data[processed_data['pressure']==1]
```

## Data Preprocessing

```python
# input data
input_data_wop = processed_data_wop.drop(['T2','T2i'], axis=1)
input_data_wop = input_data_wop.iloc[:,:-2]
targets_wop = processed_data_wop.iloc[:,-2:]
```

Standardlisation of the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaled_input_wop = scaler.fit_transform(input_data_wop)
```

Deploying scaler

```python
import joblib
joblib.dump(scaler, 'scaler.pkl')
```

## Split in train, test and validation data

```python
#WOP data split
from sklearn.model_selection import train_test_split
wop_train, wop_test_val, wop_targets_train, wop_target_test_val = train_test_split(scaled_input_wop, targets_wop, test_size=0.2, random_state=42)
```

```python
# spliting test and validation data
wop_test, wop_validation, wop_target_test, wop_target_validation = train_test_split(wop_test_val, wop_target_test_val, test_size=0.5) 
```

## Save datasets in *.npz

```python
np.savez('tcr_wop_train', input=wop_train, targets=wop_targets_train)
np.savez('tcr_wop_validation', input=wop_validation, targets=wop_target_validation)
np.savez('tcr_wop_test', input=wop_validation, targets=wop_target_validation)
```
